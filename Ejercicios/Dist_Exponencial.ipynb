```python
#Distribución Exponencial
#Contexto: Una estación de carga rápida para vehículos eléctricos quiere simular el tiempo entre la llegada de los vehículos para prever la cantidad
#de puntos de carga necesarios durante el día. La llegada de los vehículos sigue una distribución exponencial con una tasa de 1 llegada cada 15 minutos, en promedio.
#Objetivo: Generar tiempos entre llegadas de vehículos eléctricos durante un día de operación (12 horas).

import numpy as np
import matplotlib.pyplot as plt

lambda_llegada = 1 / 15  #llegadas = 1 cada 15 minutos
duracion_dia = 12 * 60  #12 horas en minutos

def generar_tiempo_entre_llegadas(lambda_llegada):
    u = np.random.uniform(0, 1)
    tiempo_entre_llegadas = -1 / lambda_llegada * np.log(1 - u)  
    return tiempo_entre_llegadas

tiempos_entre_llegadas = []
tiempo_total = 0
while tiempo_total < duracion_dia:
    tiempo = generar_tiempo_entre_llegadas(lambda_llegada)
    tiempos_entre_llegadas.append(tiempo)
    tiempo_total += tiempo

print("Tiempos entre llegadas generados (en minutos):", tiempos_entre_llegadas)

plt.hist(tiempos_entre_llegadas, bins=10, edgecolor='black', alpha=0.7)
plt.title('Tiempos entre Llegadas de Vehículos (Distribución Exponencial)')
plt.xlabel('Tiempo entre llegadas (minutos)')
plt.ylabel('Frecuencia')
plt.show()
```
